```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
jfo_CK = pd.read_csv('logL_obs_2020jfo_SDAUBgVriz.txt', sep = '\t', names = (['MJD', 'lum', 'dlum']))
```

```python
jfo_CK_bb = pd.read_csv('logL_bb_2020jfo_SDAUBgVriz.txt', sep = '\t', names = (['MJD', "lum", 'dlum']))
```

```python
jfo_ex = pd.read_csv('jfo_bol_LC', sep = ' ')
```

```python
jfo_ex
```

```python
fig, ax = plt.subplots(2, figsize = (10,10))
ax[0].errorbar(jfo_CK['MJD'], jfo_CK['lum'], yerr = jfo_CK['dlum'], label = 'CK bol LC', marker = 'o')
ax[1].errorbar(jfo_ex['Time (MJD)'].astype('float'), np.log10(jfo_ex['Log10(Bol. Lum)'].astype('float')), 
             yerr = (jfo_ex['Log10(Bol. Err)'].astype('float') / (jfo_ex['Log10(Bol. Lum)']).astype('float')), marker = 'o')
ax[0].set_xlabel('Time (MJD)')
ax[1].set_xlabel('Time (days from peak)')
fig.supylabel('log(Luminosity)')
```

```python
plt.figure(figsize=(10,10))
plt.errorbar(jfo_CK['MJD'] - 58975.25, jfo_CK['lum'], yerr = jfo_CK['dlum'], marker = 'o', label = 'CK')
plt.errorbar(jfo_ex['Time (MJD)'].astype('float'), np.log10(jfo_ex['Log10(Bol. Lum)'].astype('float')), 
             yerr = (jfo_ex['Log10(Bol. Err)'].astype('float') / (jfo_ex['Log10(Bol. Lum)']).astype('float')), marker = 'o', color = 'r',
             label = 'extrabol')
plt.xlabel('Time (days from peak)')
plt.ylabel('log Lum')
plt.legend()
```

```python
plt.figure(figsize=(10,10))
plt.errorbar(jfo_CK_bb['MJD'] - 58975.25, jfo_CK_bb['lum'], yerr = jfo_CK_bb['dlum'], marker = 'o', label = 'CK_bb')
plt.errorbar(jfo_ex['Time (MJD)'].astype('float'), np.log10(jfo_ex['Log10(Bol. Lum)'].astype('float')), 
             yerr = (jfo_ex['Log10(Bol. Err)'].astype('float') / (jfo_ex['Log10(Bol. Lum)']).astype('float')), marker = 'o', color = 'r',
             label = 'extrabol')
plt.xlabel('Time (days from peak)')
plt.ylabel('log Lum')
plt.legend()
```
