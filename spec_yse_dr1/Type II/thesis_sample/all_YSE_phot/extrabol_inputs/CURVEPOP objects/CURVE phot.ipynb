```python
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
file = "SN 2004A_ another Type II-P supernova with a red supergiant progenitor _ Oxford Academic.html"
```

```python
with open(file, 'r') as f:
    data = f.read()
S = BeautifulSoup(data, 'html.parser')
tabdata = S.find('table')
print(tabdata)
# The first tr contains the field names.
headings = [th.get_text().strip() for th in tabdata.find("tr").find_all("th")]
print(headings)

rows = []
for row in tabdata.find_all("tr"):
    datasets = [dataset.text.strip() for dataset in row.find_all('td')]
    if datasets:
        rows.append(datasets)
```

```python
sn04A = pd.DataFrame(rows, columns=headings)
sn04A.columns = [c.replace('\n', ' ') for c in sn04A.columns]
sn04A.columns = [c.replace('.', ' ') for c in sn04A.columns]


sn04A
```

```python
sn04A.dtypes
```

```python
def get_errs(column):
    mag = column.str.extract(r'(\d+(\.\d+)?)')[0]
    err = column.str.extract(r'\((\d+(\.\d+)?)\)')[0]
    return mag, err 
```

```python
Bmags, Berrs = get_errs(sn04A['B              '])
Vmags, Verrs = get_errs(sn04A['V              '])
Rmags, Rerrs = get_errs(sn04A['R              '])
Imags, Ierrs = get_errs(sn04A['I              '])
```

```python
sn04A['Bmag'] = Bmags.astype('float')
sn04A['Berr'] = Berrs.astype('float')
sn04A['Vmag'] = Vmags.astype('float')
sn04A['Verr'] = Verrs.astype('float')
sn04A['Rmag'] = Rmags.astype('float')
sn04A['Rerr'] = Rerrs.astype('float')
sn04A['Imag'] = Imags.astype('float')
sn04A['Ierr'] = Ierrs.astype('float')
```

```python
sn04A = sn04A.drop(columns=['B              ', 'V              ', 'R              ', 'I              '])
sn04A
```

```python
sn04A['Bmag']
    
```

```python
sn04A_mag_subset = ['Bmag', 'Vmag', 'Rmag', 'Imag']
sn04A_mag_df = sn04A[sn04A_mag_subset]

sn04A_err_subset = ['Berr', 'Verr', 'Rerr', 'Ierr']
sn04A_err_df = sn04A[sn04A_err_subset]
sn04A_err_df
```

```python
sn04a_mag_array = sn04A_mag_df.to_numpy()
sn04a_mag_array_concat = np.concatenate(sn04a_mag_array)
sn04a_err_array = sn04A_err_df.to_numpy()
sn04a_err_concat = np.concatenate(sn04a_err_array)
sn04a_filts = np.tile(['B', 'V', 'R', 'I'], 21)
sn04a_filts
```

```python
sn04A_no_mag = ['Date              ', 'JD (245-0000+)              ', 'Phase (d)              ', 'Telescope plus instrument              ']
sn04A_no_mag_df = sn04A[sn04A_no_mag]
sn04A_no_mag_df
```

```python
repeat_04A_vals = np.repeat(sn04A_no_mag_df.values, 4, axis=0)
repeat_04a_df = pd.DataFrame(repeat_04A_vals, columns=sn04A_no_mag_df.columns)
repeat_04a_df['MAG'] = sn04a_mag_array_concat
repeat_04a_df['ERR'] = sn04a_err_concat
repeat_04a_df['FLT'] = sn04a_filts
```

```python
md09_no_mag = ['JD-2400000.5', 'Instrument']
md09_no_mag_df = md09[md09_no_mag]
md09_repeat = np.repeat(md09_no_mag_df.values, 5, axis=0)
md09_repeat_df = pd.DataFrame(md09_repeat, columns=md09_no_mag_df.columns)
md09_repeat_df
md09_mag = ['U', 'B', 'V', 'R', 'I']
md09_err = ['Uerr', 'Berr', 'Verr', 'Rerr', 'Ierr']
md09_mag_df = md09[md09_mag]
md09_err_df = md09[md09_err]

md09_mag_array = md09_mag_df.to_numpy()
md09_err_array = md09_err_df.to_numpy()

md09_mag_concat = np.concatenate(md09_mag_array)
md09_err_concat = np.concatenate(md09_err_array)
md09_filts = np.tile(['U', 'B', 'V', 'R', 'I'], len(md09_mag_array))
md09_repeat_df['MAG'] = md09_mag_concat
md09_repeat_df['ERR'] = md09_err_concat
md09_repeat_df['FLT'] = md09_filts
md09_repeat_df['JD-2400000.5'] = 2400000.5 + md09_repeat_df['JD-2400000.5']
md09_repeat_df = md09_repeat_df[['JD-2400000.5', 'MAG', 'ERR', 'FLT', 'Instrument']]
md09_repeat_df = md09_repeat_df.rename(columns={'JD-2400000.5':'MJD', 'Instrument':'Inst'})

not_SVO = {'U':'NOT/ALFOSC.Bes_U', 'B':'NOT/ALFOSC.Bes_B', 'V':'NOT/ALFOSC.Bes_V', 'R':'NOT/ALFOSC.Bes_R', 'I':'NOT/ALFOSC.Bes_I'}
lt_SVO = {'U':'Liverpool/RATCam.sdss-u','B':'Liverpool/RATCam.bessel-B', 'V':'Liverpool/RATCam.bessel-V', 'R':'Liverpool/RATCam.sdss-r', 'I':'Liverpool/RATCam.sdss-i'}
lco_SVO = {'B':'LasCumbres/LasCumbres.Bessel_B', 'V':'LasCumbres/LasCumbres.Bessel_V', 'R':'LasCumbres/LasCumbres.Bessel_R', 'I':'LasCumbres/LasCumbres.Bessel_I'}
efosc_SVO = {'U':'LaSilla/EFOSC.ESO640','B':'LaSilla/EFOSC.ESO639', 'V':'LaSilla/EFOSC.ESO641', 'R':'LaSilla/EFOSC.ESO642', 'I':'LaSilla/EFOSC.ESO705'}
calto_SVO = {'B':'CAHA/CAHA.448_75ab', 'V':'CAHA/CAHA.534_97b', 'R':'CAHA/CAHA.641_158', 'I':'CAHA/CAHA.850_150ab'}
afosc_SVO = {'B':'LaSilla/DFOSC.Bessel_B', 'V':'LaSilla/DFOSC.Bessel_V', 'R':'LaSilla/DFOSC.Bessel_B', 'I':'LaSilla/DFOSC.Gunn_i'}
gen_SVO = {'B':'Generic/Johnson.B', 'V':'Generic/Johnson.V', 'R':'Generic/Johnson.R', 'I':'Generic/Johnson.I'}
md09_repeat_df = md09_repeat_df.dropna()

dict_list = [('NOT-(ALFOSC)', not_SVO),('LT-(RATCam)', lt_SVO),('FTN-(EM01)', lco_SVO),('NTT-(EFOSC)', efosc_SVO),
('CALTO-(CAFOS)', calto_SVO), ('Mt.Ekar1.82m-(AFOSC)', afosc_SVO), ('WISE-(PI-CCD)', gen_SVO), ('WISE-(LAIWO)', gen_SVO),
('WISE-(INDEF)', gen_SVO)]

# Function to apply mappings
def apply_mappings(row):
    for ins, mapping in dict_list:
        if row['Inst'] == ins:
            return mapping.get(row['FLT'], row['FLT'])
    return row['FLT']  

md09_repeat_df['FLT'] = md09_repeat_df.apply(apply_mappings, axis=1)

md09_repeat_df['SYS'] = 'AB'
md09_repeat_df = md09_repeat_df.drop(columns='Inst')
```

```python
a12_griz = pd.read_csv('2012A_griz.txt', sep=' ')
a12_griz_SVO = {'g':'SLOAN/SDSS.g', 'r':'SLOAN/SDSS.r', 'i':'SLOAN/SDSS.i', 'z':'SLOAN/SDSS.z'}
```

```python
A12_ubvri_mjd_df = A12_ubvri['MJD']
a12_griz_mjd_df = a12_griz['MJD']

A12_ubvri_mjd = np.repeat(A12_ubvri_mjd_df.values, 5, axis=0)
a12_griz_mjd = np.repeat(a12_griz_mjd_df.values, 5, axis=0)
a12_ubvri_mag = ['U', 'B', 'V', 'R', 'I']
a12_griz_mag = ['u', 'g', 'r', 'i', 'z']
a12_ubvri_err = ['Uerr', 'Berr', 'Verr', 'Rerr', 'Ierr']
a12_griz_err = ['uerr', 'gerr', 'rerr', 'ierr', 'zerr']
a12_ubvri_mag_df = A12_ubvri[a12_ubvri_mag]
a12_griz_mag_df = a12_griz[a12_griz_mag]
a12_ubvri_err_df = A12_ubvri[a12_ubvri_err]
a12_griz_err_df = a12_griz[a12_griz_err]
a12_ubvri_mag_array = a12_ubvri_mag_df.to_numpy()
a12_griz_mag_array = a12_griz_mag_df.to_numpy()
a12_ubvri_err_array = a12_ubvri_err_df.to_numpy()
a12_griz_err_array = a12_griz_err_df.to_numpy()

a12_ubvri_mag_concat = np.concatenate(a12_ubvri_mag_array)
a12_griz_mag_concat = np.concatenate(a12_griz_mag_array)
a12_ubvri_err_concat = np.concatenate(a12_ubvri_err_array)
a12_griz_err_concat = np.concatenate(a12_griz_err_array)
a12_ubvri_filts = np.tile(['U', 'B', 'V', 'R', 'I'], len(a12_ubvri_mag_array))
a12_griz_filts = np.tile(['u', 'g', 'r', 'i', 'z'], len(a12_griz_mag_array))
repeat_a12_ubvri_df = pd.DataFrame(A12_ubvri_mjd, columns=['MJD'])
repeat_a12_griz_df = pd.DataFrame(a12_griz_mjd, columns=['MJD'])
repeat_a12_ubvri_df['MAG'] = a12_ubvri_mag_concat
repeat_a12_griz_df['MAG'] = a12_griz_mag_concat
repeat_a12_ubvri_df['ERR'] = a12_ubvri_err_concat
repeat_a12_griz_df['ERR'] = a12_griz_err_concat
repeat_a12_ubvri_df['FLT'] = a12_ubvri_filts
repeat_a12_ubvri_df['FLT'] = repeat_a12_ubvri_df['FLT'].map(A12_ubvri_svo)
repeat_a12_griz_df['FLT'] = a12_griz_filts
repeat_a12_griz_df['FLT'] = repeat_a12_griz_df['FLT'].map(a12_griz_SVO)
repeat_a12_ubvri_df['SYS'] = 'AB'
repeat_a12_griz_df['SYS'] = 'AB'

a12_all = pd.concat([repeat_a12_ubvri_df, repeat_a12_griz_df], axis=0).reset_index(drop=True)
a12_all = a12_all.dropna()
a12_all.to_csv('2012A.dat', header=False, index=False, sep = ' ')
```
