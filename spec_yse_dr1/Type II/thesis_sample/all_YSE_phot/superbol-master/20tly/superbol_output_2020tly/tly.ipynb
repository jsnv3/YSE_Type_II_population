```python
import numpy as np 
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
table = Table.read('logL-obs_2020tly_gcwroiz.txt', format = 'ascii', names = ('MJD', 'lum', 'dlum'))
```

```python
plt.errorbar(table['MJD'], table['lum'], yerr=table['dlum'])
plt.title('2020tly Bol. Lum.')
plt.xlabel('MJD')
plt.ylabel('Luminosity')
```

```python
times = interp_lc['#phase']
plt.figure(figsize=(10,11))
plt.errorbar(times, interp_lc['g'], interp_lc['err'], label = 'g', marker = 'o', color = 'green')
plt.errorbar(times, interp_lc['c'], interp_lc['err_1'], label = 'c-ATLAS', marker = 'o', color = 'cyan')
plt.errorbar(times, interp_lc['w'], interp_lc['err_2'], label = 'w', marker = 'o', color = 'gray')
plt.errorbar(times, interp_lc['r'], interp_lc['err_3'], label = 'r', marker = 'o', color = 'r')
plt.errorbar(times, interp_lc['o'], interp_lc['err_4'], label = 'o-ATLAS', marker = 'o', color = 'orange')
plt.errorbar(times, interp_lc['i'], interp_lc['err_5'], label = 'i', marker = 'o', color = 'gold')
plt.errorbar(times, interp_lc['z'], interp_lc['err_6'], label = 'z', marker = 'o', color = 'k')
plt.gca().invert_yaxis()
plt.xlabel('MJD')
plt.ylabel('App Mag')
plt.legend(ncol = 3)
plt.title('2020tly, ref = g')
```
