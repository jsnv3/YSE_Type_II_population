```python
import numpy as np 
import matplotlib.pyplot as plt 
from astropy.table import Table
```

```python
table = Table.read('logL-obs_2020jfo_SDAuBgVriz.txt', format = 'ascii', names = ('MJD', 'lum', 'dlum'))
```

```python
plt.errorbar(table['MJD'], table['lum'], table['dlum'])
plt.title('2020jfo Bol. Lum.')
plt.xlabel('MJD')
plt.ylabel('Luminosity')
```

```python
interp_lc = Table.read('interpolated-lcs_2020jfo_SDAuBgVriz.txt', format = 'ascii')
```

```python
interp_lc
```

```python
times = interp_lc['#phase']
```

```python
plt.figure(figsize = (10,11))
plt.errorbar(times, interp_lc['S'], interp_lc['err'], marker = 'o', label = 'UVW2', color = 'purple')
plt.errorbar(times, interp_lc['D'], interp_lc['err_1'], label = 'UVM2', marker = 'o', color = 'pink')
plt.errorbar(times, interp_lc['A'], interp_lc['err_2'], label = 'UVM1', marker = 'o')
plt.errorbar(times, interp_lc['u'], interp_lc['err_3'], label = 'U', marker = 'o', color = 'violet')
plt.errorbar(times, interp_lc['B'], interp_lc['err_4'], label = 'B', marker = 'o', color = 'blue')
plt.errorbar(times, interp_lc['g'], interp_lc['err_5'], label = 'g', marker = 'o', color = 'green')
plt.errorbar(times, interp_lc['V'], interp_lc['err_6'], label = 'V', marker = 'o')
plt.errorbar(times, interp_lc['r'], interp_lc['err_7'], label = 'r', marker = 'o', color = 'red')
plt.errorbar(times, interp_lc['i'], interp_lc['err_8'], label = 'i', marker = 'o', color = 'gold')
plt.errorbar(times, interp_lc['z'], interp_lc['err_9'], label = 'z', marker = 'o', color = 'k')
plt.gca().invert_yaxis()
plt.legend(ncol = 5)
plt.xlabel('MJD')
plt.ylabel('App Mag')
plt.title('2020jfo, ref = r')
```
